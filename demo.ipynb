HISTOGRAM


```python
# import matplotlib.pyplot as plt
import seaborn as sns

# Optional: apply seaborn style for professional look
sns.set_style("whitegrid")

plt.figure(figsize=(8, 5))

plt.hist(
    stud['writing score'],
    bins=14,
    color="#BEFFC4",    
    edgecolor="#1210107A",   
    linewidth=1.2,
    alpha=0.9
)

plt.xlabel('writing Score', fontsize=12, fontweight='bold')
plt.ylabel('Frequency', fontsize=12, fontweight='bold')
plt.title('Distribution of Student writing Scores',
          fontsize=14, fontweight='bold', pad=12)

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))

ti['Pclass'].value_counts().plot(kind='pie',autopct='%1.2f%%',startangle=90,colors=["#F50014", "#00FF3C", "#009DFF"],wedgeprops={'edgecolor': 'white', 'linewidth': 1.5})

plt.title('Passenger Class Distribution',fontsize=14,fontweight='bold',pad=1)

plt.ylabel('')  
plt.tight_layout()
plt.show()
```
