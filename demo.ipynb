HISTOGRAM


```python
# import matplotlib.pyplot as plt
import seaborn as sns

# Optional: apply seaborn style for professional look
sns.set_style("whitegrid")

plt.figure(figsize=(8, 5))

plt.hist(
    stud['writing score'],
    bins=14,
    color="#BEFFC4",    
    edgecolor="#1210107A",   
    linewidth=1.2,
    alpha=0.9
)

plt.xlabel('writing Score', fontsize=12, fontweight='bold')
plt.ylabel('Frequency', fontsize=12, fontweight='bold')
plt.title('Distribution of Student writing Scores',fontsize=14, fontweight='bold', pad=12)

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

scatter plot


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

sns.scatterplot(data=stud, x='math score', y='writing score',
                color='gray', s=60, alpha=0.6)

sns.regplot(data=stud, x='math score', y='writing score',
            scatter=False, color='blue', label='Math')

sns.regplot(data=stud, x='writing score', y='math score',
            scatter=False, color='red', label='Writing')

plt.title('Math vs Writing Scores')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

sns.scatterplot(data=stud, x='math score', y='reading score',hue='gender',
                color='gray', s=60, alpha=0.6)

sns.regplot(data=stud, x='math score', y='reading score',
            scatter=False, color="#F8DB05", label='Math')

sns.regplot(data=stud, x='reading score', y='math score',
            scatter=False, color='#000000', label='reading')

plt.title('Math vs reading Scores')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

sns.scatterplot(data=stud, x='reading score', y='writing score',hue='gender',
                color='gray', s=60, alpha=0.6)

sns.regplot(data=stud, x='reading score', y='writing score',
            scatter=False, color="#9EFF03", label='Math')

sns.regplot(data=stud, x='writing score', y='reading score',
            scatter=False, color='#1C4D8D', label='Writing')

plt.title('reading vs Writing Scores')
plt.legend()
plt.show()
```
