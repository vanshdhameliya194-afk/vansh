HISTOGRAM


```python
import matplotlib.pyplot as plt

plt.hist(stud['math score'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.title('STUDENT Math Scores')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Use sns.histplot, not plt.histplot
plt.hist(stud['writing score'], bins=10, edgecolor='black',color='yellow')

plt.xlabel('read')
plt.ylabel('Frequency')
plt.title('stud read Distribution')
plt.show()
```
